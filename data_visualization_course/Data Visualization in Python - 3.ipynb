## Data Visualization with Altair

### Introduction to Altair

Altair is a Python library designed for statistical visualization. It is based on Vega and Vega-Lite, which are both visualization grammar that allows you to describe the visual appearance and interactive behavior of a visualization in a JSON format. ***Today we learn to prepare Line Plot and Stacked Area Chart with Altair package.*** Let's start.

### Creating Line Plot with Altair

```python
import pandas as pd
import altair as alt

print("Loading Completed")
```

In Altair, we start by importing the necessary packages first, ***as we have done with all our visuals so far.*** This time, our required packages are Pandas for loading the dataset and Altair for visualization.

```python
df = pd.read_csv("Ice Cream Sales.csv")
print(df)
```

***As always, after importing the necessary packages, it's time to load the dataset.*** Using Pandas's read_csv method, we have loaded the csv file that was previously on our computer.

```python
import pandas as pd
import altair as alt

df = pd.read_csv("Ice Cream Sales.csv")

alt.Chart(df).mark_line().encode(
    x = 'Date',
    y = 'Chocolate',
)
```

And yes, as you can see, ***with a short code, we got the Line Plot in the Altair package, although it has some issues.*** Let's take a look at what's in this code. In our code, ***the mark_line method states that we want to create Line Plot. Inside the parentheses of the alt.chart method, we determine the values of the x and y axes.*** And after this steps, we see the first visual we created in Altair. Let's start customizing this visual now.

```python
import pandas as pd
import altair as alt

df = pd.read_csv("Ice Cream Sales.csv")

alt.Chart(df).mark_line(color = "red", opacity = 0.3, point = True).encode(
    x = 'Date:T',
    y = 'Chocolate'
)
```

***We made our customizations in Altair, let's take a look at the customizations we made and how we do these customizations.*** We started the first changes in the brackets of mark_line. By specifying parameters such as color, opacity and point in mark_line, we added features such as color, transparency and punctuation of values to our Line Plot. The change after the changes of mark.line is T, next to Date. In our previous visual, the values on the x-axis were arranged in alphabetical order and it was a wrong visual. We have corrected the alignment of the x axis using the Date: T method.

In Altair, we first prepared our Line Plot, then customized it, and we have fully prepared our Line Plot in the Altair package. ***Now it is time to prepare Stacked Area Chart with Altair.***

### Creating Stacked Area Chart with Altair

In today's lesson, next is the preparation of Stacked Area Chart with Altair. ***Before we start preparing our visual, let's take a look at what is Stacked Area Chart.*** Stacked Area Graphs work in the same way as simple Area Graphs do, except for the use of multiple data series that start each point from the point left by the previous data series. The entire graph represents the total of all the data plotted. Stacked Area Graphs also use the areas to convey whole numbers, so they do not work for negative values. Overall, they are useful for comparing multiple variables changing over an interval.

***Now that we've learned what Stacked Area Chart is, we can move on to learn how to do it.***

```python
import altair as alt
from vega_datasets import data
```

As always, we first import the necessary packages.

Normally, in every data visualization package, we would load the dataset after importing the required packages, but this time we don't load the dataset. ***For Stacked Area Chart, we will use one of Altair's ready-made datasets.***

```python
import altair as alt
from vega_datasets import data

source = data.iowa_electricity()

alt.Chart(source).mark_area().encode(
    x = "year",
    y = alt.Y("net_generation", stack="normalize"),
    color = "source"
)
```

And this is Stacked Area Chart. Now let's check how we prepared the Stacked Area Chart in Altair, line by line. As you will notice, we also imported a package named data while importing the packages. This package was imported for us to use the ready-made datasets in Altair. While preparing our visual, we named it as source to use iowa_electricity dataset from within the ready-made data. And then we prepared our visual similar to what we did in Line Plot. ***The main thing to note here is that we use mark_area instead of mark_line.***

Thus, in the first part of our 3rd Lesson, we learned to make Line Plot and Stacked Area Chart using Altair data visualization package. ***In the rest of this lesson, we will learn how to visualize data with ggplot. Let's move on to ggplot.***

## Data Visualization with ggplot

### Introduction to ggplot

ggplot, like Python's other data visualization packages, is a package that allows us to do various visualizations with its own syntax. ggplot comes from ggplot2 at R. Now that we know what ggplot is, we can move on to how to use ggplot. ***In the rest of the lesson, we will learn to prepare Line Plot and Histogram using the ggplot.*** Let's start.

### Creating Line Plot with ggplot

```python
import pandas as pd
from plotnine import *
from plotnine.data import *

print("Loading Completed")
```

As always, we start by importing our necessary packages. ***You might think why plotnine was imported instead of ggplot here. With the release of the latest version of Pandas, ggplot is no longer functional, so we decided to use plotnine, which is from the same family as ggplot (you will see it in the code), as an alternative to ggplot.*** As usual, pandas was imported to load the dataset and to plotnine for visualization. The reason why plotnine.data is imported is that in this plot we will use one of the plotnine's ready datasets, not our own data.

```python
print(economics.head())
```

As we have just mentioned, since we are using one of the plotnine's ready-made datasets, we do not need to load external datasets with Pandas, we can switch directly to printing the dataset on the screen. ***After reviewing our data, it's time to prepare our Line Plot.***

```python
import pandas as pd
from plotnine import *
from plotnine.data import *

(
    ggplot(economics, aes(x = 'date', y = 'uempmed'))
    + geom_line()
)
```

And yes, here is a Line Plot prepared with plotnine (ggplot). Let's look at how we prepared the Line Plot. ***First, we have determined our dataset with economics and which columns we will use with aes in the brackets of ggplot. After this process, we determined the type of data visualization we want to prepare by typing geom_line.*** And our visual has been prepared. Now we can start customizing for a better visual.

```python
import pandas as pd
from plotnine import *
from plotnine.data import *

(
    ggplot(economics, aes(x = 'date', y = 'uempmed'))
    + geom_line(color = "blue", size = 1)
    + labs(x = 'Date', y = 'Median Duration of Unemployment')
)
```

With the above transactions, we have also completed our arrangements. ***While customizing the Line Plot we prepared with Plotnine, we first determined the color and size of our line with geom_line. Then we made new names to the x and y axes with labs.*** Thus, our visual has been prepared.

Thus, we have completed our Line Plot preparation section with ggplot. ***Next is the Histogram preparation with ggplot, which will be the last visual of our lesson.*** Let's go.

### Creating Histogram with ggplot

Before we start preparing Histograms with ggplot, let's have a look at what is histogram. ***A Histogram is an approximate representation of the distribution of numerical data.*** Now that we know the Histogram, we can move on to preparing Histogram with ggplot.

```python
import pandas as pd
from plotnine import *
from plotnine.data import *

print("Loading Completed")
```

We started our process by importing the necessary packages. ***This time, we have imported the plotnine.data package as we will be using one of plotnine's own ready-made datasets.*** Now we can move on to dataset.

```python
print(diamonds.head(5))
```

***Since we are using ready-made dataset, we have printed the diamond dataset we will use this time on the screen without loading dataset with Pandas.***  Now it's time to prepare the Histogram

```python
import pandas as pd
from plotnine import *
from plotnine.data import *

(
    ggplot(diamonds, aes(x='carat'))
    + geom_histogram()
)
```

As you can see, we have obtained Histogram in ggplot with a short code. ***First, we determined the dataset we will use in ggplot and the x value we will use with aes. Then we stated that the plot we tried to prepare with the geom_histogram method is a histogram.*** And here is a Histogram made with ggplot.

```python
import pandas as pd
from plotnine import *
from plotnine.data import *

(
    ggplot(diamonds, aes(x='carat'))
    + geom_histogram(binwidth = 0.70, bins = 5)
)
```

After we have prepared our Histogram, we can start editing it. In the editing of our Histogram, we can determine the width of bin with binwidth and the number of bin with bins in the geom_histogram.
