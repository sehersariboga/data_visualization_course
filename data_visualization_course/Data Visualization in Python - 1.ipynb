After loading our packages, ***it is time to load our dataset.*** We use Pandas to load our dataset in the easiest way possible. ***When loading our dataset with the pd.read_csv function, we call this data set df.*** After loading our dataset, we can print our dataset on the screen to check the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ice Cream Sales.csv")

fig, ax = plt.subplots()

plt.show()
```

Using fig, ax = plt.subplots () helps us a lot when we are preparing multiple images or comparing images. So it is always most efficient to use the fig, ax method. If we leave the brackets of plt.subplots empty in the fig, ax method, it creates an empty and one visual. With the plt.show function, we enable our visual to be displayed. ***Fig, ax and plt.show are commonly used in matplotlib for each data visualization package.***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ice Cream Sales.csv")

fig, ax = plt.subplots()

ax.plot(df["Date"], df["Chocolate"])

plt.show()
```

And yes, at the end of a long routine steps and introduction, we have our first visual. ***Although it needs a lot of editing, the visual we get can be called a line plot.*** Now let's see how we got this visual. In Matplotlib, we can use the ***"plt"*** method to obtain a line plot. When using the ax.plt method, the first value we put in parentheses represents the x-axis and the second value represents the y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ice Cream Sales.csv")

fig, ax = plt.subplots()

ax.plot(df["Date"], df["Chocolate"])

ax.set_xticklabels(df["Date"], rotation = 90)
ax.set_xlabel("Date")
ax.set_ylabel("Sales")
ax.set_title("Chocolate Sales")

plt.show()
```

After arranging the x and y axes, we have reached our first clear visual. Here, ***we gave the name of the x-axis with the ax.set_xlabel method and the name of the y-axis with the ax.set_ylabel method.*** After naming the X and Y axes, ***we put the title of our visual with the ax.set_title method.*** 

More importantly, ***we made the labels readable by using the rotation value in the ax.set_xticklabels method and rotating the labels on the x axis.*** Of course, we can make changes to the labels on the y-axis using the ax.set_yticklabels method if we want.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ice Cream Sales.csv")

fig, ax = plt.subplots()

ax.plot(df["Date"], df["Chocolate"], color = "r", marker = "o", label = "Chocolate" )
ax.plot(df["Date"], df["Vanilla"], color = "b", marker = "d", label = "Vanilla" )

ax.set_xticklabels(df["Date"], rotation = 90)
ax.set_xlabel("Date")
ax.set_ylabel("Sales")
ax.set_title("Ice Cream Sales")
ax.legend()

plt.show()
```

Finally, we compared two different ice cream flavors using the ax, plt method. We also used new customizations while making these comparisons. As you can see, ***these customizations are color, marker and label.*** Color determines the color of our line, marker determines the place corresponding to each value and label determines the name of our line. Finally, we added a legend to our visual by writing the ax.legend () method under our code.

As you can see above, we have completed our first visual. Of course, we can customize a lot more when preparing a line plot, but this image can be considered sufficient to prepare a legible visual. ***At this stage of our lesson, we prepared Line Plot using Matplotlib package, and next is to prepare Scatter Plot with Matplotlib package.*** Let's create a Scatter Plot now.

### Creating Scatter Plot with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

plt.show()
```

While preparing a scatter plot with Matplotlib package, ***we first perform the routine steps we do*** for each data visualization type while preparing visuals in Python. These routine steps are to ***import Matplotlib and Pandas packages, load the dataset and prepare the method fig, ax = plt.subplots (), respectively.***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ice Cream Sales.csv")

fig, ax = plt.subplots()

ax.scatter(df["Date"], df["Chocolate"])

plt.show()
```

As we can see above, we can obtain a scatter plot by using the ***ax.scatter method instead of the ax.plot method*** we use when preparing Line Plot. Although the scatter plot needs to be edited as much as the line plot in the first step, we can call the visual scatter plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ice Cream Sales.csv")

fig, ax = plt.subplots()

ax.scatter(df["Date"], df["Chocolate"])

ax.set_xticklabels(df["Date"], rotation = 90)
ax.set_xlabel("Date")
ax.set_ylabel("Sales")
ax.set_title("Chocolate Sales")

plt.show()
```

While customizations are done, we can apply the same x and y axis arrangements that we first made in line plot to the scatter plot. To repeat these, ***we create the name of the x label with ax.set_xlabel, the name of the y label with ax.set_ylabel, and the title of our plot with the ax.set_title method.***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ice Cream Sales.csv")

fig, ax = plt.subplots()

ax.scatter(df["Date"], df["Chocolate"], c = "r", s = 10, alpha = 0.3, label = "Chocolate")
ax.scatter(df["Date"], df["Vanilla"], c = "b", s = 17.3, alpha = 0.7, label = "Vanilla")

ax.set_xticklabels(df["Date"], rotation = 90)
ax.set_xlabel("Date")
ax.set_ylabel("Sales")
ax.set_title("Ice Cream Sales")
ax.legend()

plt.show()
```

After editing the x and y axis, we can move on to the ***scatter plot specific arrangements.*** If we look at the arrangements specific to the scatter plot in general, ***we have determined the color with "color", the size of our point with "s", and the transparency of our point with "alpha".*** As you can see in our code, Vanilla's points are bigger and darker because Vanilla's "s" and "alpha" values are larger than Chocolate.

In this part of our lesson and in the first half hour, we learned how to make Line Plot and Scatter Plot using Matplotlib package. In general, we saw that ***we can prepare our visuals using ax.plot in Line Plot and ax.scatter method in Scatter Plot.*** For the rest of the lesson, we'll cover Seaborn, another widely used data visualization package from Python.

### Creating Line Plot with Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Loading Completed")
```

Like Matplotlib, ***there are packages that we need to import while visualizing in Seaborn.*** These are Pandas for dataset loading, Seaborn and Matplotlib for visualization. The reason why we import both Seaborn and Matplotlib instead of importing only Seaborn for visualization is that ***Matplotlib is package that Seaborn built on top of.***

```python
df = pd.read_csv("Ice Cream Sales.csv")
print(df)
```

After installing our packages, we can also upload our dataset using the Pandas package, ***like before.***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Ice Cream Sales.csv")

sns.relplot(x = "Date", y = "Chocolate", data = df, kind = "line")

plt.show()
```

As we have seen above, we can obtain a line plot using the sns.relplot method, although it needs a lot of editing. So what is relplot? There are generally two types of data visualization, according to Seaborn. ***These are relplot and catplot. Relplot represents the relational plot; catplot represents the categorical plot.*** As you can see from the name, relplot is used when comparing the relationships between two values, so relplot is used in cases such as scatter plot and line plot. Catplot, on the other hand, is usually used in categorical situations where there is no comparison, such as bar plot.

So how to use relplot? When using Relplot, the x inside the parentheses represents the x-axis and y represents the y-axis. Apart from these, data indicates which data we use (df in this case) and kind indicates which type of data visualization we use (line in this case).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Ice Cream Sales.csv")

sns.relplot(x = "Date", y = "Chocolate", data = df, kind = "line")

plt.xticks(rotation = 90)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Chocolate Sales")

plt.show()
```

After preparing a simple visual as we did when preparing our previous visuals, we first arrange the x and y axes. ***Here, we have written the x label with plt.xlabel, the y label with plt.ylabel, and the title with plt.title. By using rotation with the plt.xticks method, we rotate the x values***, that is the dates, 60 degrees so that they can be read.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Ice Cream Sales.csv")

sns.relplot(x = "Date", y = "Chocolate", data = df, sort = False, kind = "line", color = "r", marker = "o", label = "Chocolate")

plt.xticks(rotation = 90)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Chocolate Sales")
plt.legend()

plt.show()
```

While preparing line plot in Seaborn, we can do the same customizations we did in matplotlib. These customizations that you will remember from Matplotlib are ***color, marker and label.***

But here, as you will notice, there has been another change. The shape of our Line Plot has changed, now we can look at why this change occurred. ***As our careful audience will see, we have added another parameter called "sort" to our code. The "sort" parameter helped us sort the values on the x-axis as we want. Before adding the "sort" parameter, our dates on the x axis were arranged in alphabetical order, after adding "sort", they were arranged as we wanted.*** Do not forget to check your x-axis while preparing a Line Plot in Seaborn, and if our sequence is wrong, do not forget to correct it with "sort".

At this stage of our lesson, we prepared Line Plot with Seaborn, one of the most common data visualization package in Python. Next is the preparation of Bar Plot with Seaborn, the last visual of this lesson.

### Creating Bar Plot with Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Ice Cream Sales.csv")
print(df)
```

Before we start preparing a Bar Plot in Seaborn, ***we take routine steps as we do in every visual.*** These routine steps are to importing the packages and loading the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Ice Cream Sales.csv")

sns.catplot(x = "Date", y = "Chocolate", data = df, kind = "bar" )

plt.show()
```

While preparing a Bar Plot at Seaborn, at first we are preparing our visual that needs editing, as in our previous visuals. We used the sns.catplot method while preparing this image. ***As we explained in our previous image, catplot is used for categorical data visualization types such as bar plot. Values in catplot are exactly the same as relplot.***  x represents the x-axis; y represents the y axis, data represents the data we load, and kind represents our data visualization type, bar for this time.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Ice Cream Sales.csv")

sns.catplot(x = "Date", y = "Chocolate", data = df, kind = "bar")

plt.xticks(rotation = 90)
plt.xlabel("Date")
plt.ylabel("Chocolate")
plt.title("Chocolate Sales")

plt.show()
```

In our visual, we make the first arrangements on the x and y axes. We have made our visual ***more readable by making the same changes we made in our previous visual.***
